```python
category_stats = df.groupby('Category')['WINNING STATUS'].agg(['sum','count'])
category_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
category_stats['Win_Rate'] = category_stats['Wins'] / category_stats['Total']
print("\nCategory-level win stats:\n", category_stats)
```

```python
tech_stats = df.groupby('Technology_Bucket')['WINNING STATUS'].agg(['sum','count'])
tech_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
tech_stats['Win_Rate'] = tech_stats['Wins'] / tech_stats['Total']
print("\nTechnology bucket stats:\n", tech_stats)
```

```python
dept_stats = df.groupby('Department')['WINNING STATUS'].agg(['sum','count'])
dept_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
dept_stats['Win_Rate'] = dept_stats['Wins'] / dept_stats['Total']
print("\nDepartment-level stats:\n", dept_stats)
```

```python
org_stats = df.groupby('Organisation')['WINNING STATUS'].agg(['sum','count'])
org_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
org_stats['Win_Rate'] = org_stats['Wins'] / org_stats['Total']
print("\nOrganisation-level stats:\n", org_stats)
```

```python
ministry_tech_stats = df.groupby(['MINISTRY/GOVT DEPT/ORGANISATION','Technology_Bucket'])['WINNING STATUS'].agg(['sum','count'])
ministry_tech_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
ministry_tech_stats['Win_Rate'] = ministry_tech_stats['Wins'] / ministry_tech_stats['Total']
print("\nMinistry + Technology stats:\n", ministry_tech_stats)
```

```python
id_stats = df.groupby('ID')['WINNING STATUS'].agg(['sum','count'])
id_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
id_stats['Win_Rate'] = id_stats['Wins'] / id_stats['Total']
consistent_ps = id_stats[id_stats['Win_Rate'] == 1]
print("\nPS with 100% win rate:\n", consistent_ps)
```

```python
ps_stats = software_df.groupby('ID')['WINNING STATUS'].agg(['sum','count'])
ps_stats.rename(columns={'sum':'Wins','count':'Total'}, inplace=True)
ps_stats['Win_Rate'] = ps_stats['Wins'] / ps_stats['Total']

# Step 2: Add metadata (Category, Tech, Dept, Org, Ministry)
meta = software_df.drop_duplicates(subset=['ID'])[['ID','Category','Technology_Bucket',
                                                   'Department','Organisation',
                                                   'MINISTRY/GOVT DEPT/ORGANISATION']]
ps_summary = ps_stats.merge(meta, on='ID', how='left')

# Step 3: Scoring formula (adjust weights if needed)
ps_summary['Score'] = (0.5*ps_summary['Win_Rate'] +
                       0.3*(ps_summary['Wins']/ps_summary['Wins'].max()) +
                       0.2*(ps_summary['Total']/ps_summary['Total'].max()))

# Step 4: Sort by Score
ps_summary_sorted = ps_summary.sort_values(by='Score', ascending=False)

# Show final summary (Software only)
ps_summary_sorted.head(50)
```
